```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Network analysis data V2.1.xlsx')
```

```python
# A lot of duplicate rows which need to be deleted for now
```

```python
df2 = df.drop_duplicates()
```

```python
print(df2.head())
print(list(df2.columns.values))
# Only unique rows are left
```

```python
# The unnamed column can be dropped, it is not of any use.

df2 = df2.drop(labels= "Unnamed: 2", axis = 1)
```

```python
df2
```

```python
df2.count(axis = 1)
```

```python
# Seeing which names are coming back in combination with multiple company's

datasplit = []

for index, row in df2.iterrows():
    if df2['Full name'].value_counts()[row['Full name']] > 1:
        print('Multiples', ("\t"), row['Full name'], ("\t"), row['Company name'])
        continue
    else:
        print('No multiples', ("\t"), row['Full name'], ("\t"), row['Company name'])
        continue
```

```python
datasplit = pd.DataFrame(datasplit)
```

```python
datasplit.head()
```

```python
multiples = datasplit[datasplit['Amount'] == 'Multiples']
no_multiples = datasplit[datasplit['Amount'] == 'No multiples']
```

```python
with pd.ExcelWriter('Processed ND2.xlsx') as writer:
    multiples.to_excel(writer,sheet_name = 'Multiples')
    no_multiples.to_excel(writer,sheet_name = 'No multiples')
```
